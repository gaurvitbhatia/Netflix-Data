# Data Cleaning and Processing

```python
flix.isnull().sum() 
```

```python
flix['rating'].isnull().value_counts()
```

```python
flix[flix['rating'].isnull()]
```

```python
flix['rating'].value_counts().head()
```

```python
flix['rating'].fillna('TV-MA', inplace = True)
```

```python
flix['rating'].isnull().value_counts()
```

```python
flix[flix['date_added'].isna()]
```

```python
flix['date_added'].isna().value_counts()
```

```python
flix[flix['date_added'].isna()]
```

```python
flix.drop(columns = 'date_added', inplace = True)
```

```python
flix.drop(columns = 'show_id', inplace = True)
```

```python
flix.head()
```

```python
flix.head()
```

```python
flix['type'].value_counts()
```

```python
flix[flix['type'] == 'TV Show']['duration'].value_counts()
```

```python
flix['movie_dur'] = flix[flix['type'] == 'Movie']['duration'].apply(lambda x : x.split()[0])
```

```python
flix['no_of_seasons'] = flix[flix['type'] == 'TV Show']['duration'].apply(lambda x : x.split()[0])
```

```python
flix['no_of_seasons'].value_counts()
```

```python
flix.head(10)
```

```python
flix['movie_dur'].fillna('0', inplace = True)
```

```python
flix['no_of_seasons'].fillna('0', inplace = True)
```

```python
flix['movie_dur'] = pd.to_numeric(flix['movie_dur'])
```

```python
flix['no_of_seasons'] = pd.to_numeric(flix['no_of_seasons'])
```

```python
flix.drop(columns = 'duration', inplace = True)
```

```python
flix.head(10)
```

```python
flix['director'].value_counts().head(15)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'release_year', data = flix[flix['release_year'] >= 2000])
```

```python
sns.countplot(x = 'type', data = flix)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'rating', data = flix)
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x = 'rating',hue = 'type', data = flix)
```

```python
plt.figure(figsize=(10,4))
plt.xlim(0.5,15)
plt.ylim(0,1400)
sns.distplot(flix['no_of_seasons'], bins = 30, kde = False, color = 'brown')
```

```python
plt.figure(figsize=(10,4))
plt.xlim(20,200)
plt.ylim(0,0.015)
sns.distplot(flix['movie_dur'], bins = 30, color = 'purple')
```
